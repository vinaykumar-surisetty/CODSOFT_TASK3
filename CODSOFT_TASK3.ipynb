```python
import random

# Define options for the game
options = ["Rock", "Paper", "Scissors"]


def get_user_choice():
  """
  Prompts the user to choose rock, paper, or scissors and validates the input.
  """
  while True:
    user_choice = input("Choose Rock, Paper, or Scissors (R/P/S): ").upper()
    if user_choice in ("R", "P", "S"):
      # Convert user's choice to full word for display
      return options[ord(user_choice) - ord('R')]
    else:
      print("Invalid choice. Please try again.")


def get_computer_choice():
  """
  Generates a random choice (rock, paper, or scissors) for the computer.
  """
  return random.choice(options)


def determine_winner(user_choice, computer_choice):
  """
  Determines the winner based on the user's choice and the computer's choice.
  """
  if user_choice == computer_choice:
    return "Tie"
  elif (user_choice == "Rock" and computer_choice == "Scissors") or (
      user_choice == "Paper" and computer_choice == "Rock") or (
      user_choice == "Scissors" and computer_choice == "Paper"):
    return "You Win!"
  else:
    return "You Lose"


def play_again():
  """
  Asks the user if they want to play another round.
  """
  while True:
    choice = input("Play again? (y/n): ").lower()
    if choice in ("y", "n"):
      return choice == "y"
    else:
      print("Invalid choice. Please enter y or n.")


def main():
  # Keep track of scores (optional)
  user_score = 0
  computer_score = 0

  while True:
    # Get user and computer choices
    user_choice = get_user_choice()
    computer_choice = get_computer_choice()

    # Determine the winner and display results
    winner = determine_winner(user_choice, computer_choice)
    print(f"You chose: {user_choice}")
    print(f"Computer chose: {computer_choice}")
    print(winner)

    # Update scores (optional)
    if winner == "You Win!":
      user_score += 1
    elif winner == "You Lose":
      computer_score += 1

    # Display scores (optional)
    if user_score or computer_score:
      print(f"Your score: {user_score}")
      print(f"Computer score: {computer_score}")

    # Ask to play again
    if not play_again():
      break

  print("Thanks for playing!")


if __name__ == "__main__":
  main()
```
